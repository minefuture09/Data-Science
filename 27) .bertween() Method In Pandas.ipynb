```python
import pandas as pd 

df = pd.read_csv('Employees.csv',parse_dates=['Start Date','Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
```

```python
condition = df['Salary'].between(60000,70000)

df[condition]
```

```python
bonus = df['Bonus %'].between(2,5)
df[bonus]
```

```python
date = df['Start Date'].between('1991-01-01','1992-01-01')
df[date]
```

```python
login = df['Last Login Time'].between('8:30AM','12:00PM')

df[login]
```
