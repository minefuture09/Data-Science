```python
import pandas as pd

nba = pd.read_csv('nba.csv')

nba['Salary']=nba['Salary'].dropna(how='all')
 
```

```python
nba['Salary'] = nba['Salary'].fillna(0).astype(int)
```

```python
nba['Rank'] = nba['Salary'].rank(ascending=False).astype(int)
```

```python
nba.sort_values('Salary',ascending=False)
```
